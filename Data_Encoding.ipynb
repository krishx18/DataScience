**One Hot Encoding or Nominal Encoding**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'color':['red','blue','green','green','red','blue']})
```

```python
# df
```

```python
 # create an instance of OneHotEncoder
ohe = OneHotEncoder()
```

```python
# perform fit and transform
sparse_matrix = ohe.fit_transform(df[['color']])
```

```python
sparse_matrix
```

```python
sparse_matrix=sparse_matrix.toarray()
```

```python
# from OneHotEncoder instance we can see the new features created
ohe.get_feature_names_out()
```

```python
# Create the dataframe
ohe_df = pd.DataFrame(sparse_matrix,columns=ohe.get_feature_names_out())
```

```python
ohe_df
```

```python
# How to deal with new data
ohe.transform([['blue']]).toarray()
```

```python
new_df = pd.concat([df,ohe_df],axis=1)
new_df.head()
```

**Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Create an instance of Ordinal Encoder
oe = OrdinalEncoder()
```

```python
df_ordinal = pd.DataFrame({'size':['small','medium','large','medium','small','large']})
```

```python
# df_ordinal
```

```python
# Let's assing ranking or order
ranks = [['small','medium','large']]
```

```python
oe=OrdinalEncoder(categories=ranks)
```

```python
oe.fit_transform(df_ordinal[['size']])
```

```python
# New data
oe.transform([['medium']])
```

**Target Guided Ordinal Encoding**

```python
df = pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
mean_values = df.groupby('city')['price'].mean().to_dict()
mean_values
```

```python
df['city_encoded'] = df['city'].map(mean_values)
```
